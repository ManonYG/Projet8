### Aucune valeur de SalePrice ne semble trop extrême ; inutile donc de supprimer des individus.

```python
# DEDAWP
df_num = data.select_dtypes(include = ['float64', 'int64'])
df_num.drop(columns=["SalePrice"], inplace=True);
df_num.hist(figsize=(16, 20), bins=50, xlabelsize=8, ylabelsize=8);
```

```python
fig, axes = plt.subplots(6, 6, figsize=(15,15))

for i,el in enumerate(list(df_num.columns.values)):
    a = df_num.boxplot(el, ax=axes.flatten()[i])
```

### Les variables contenant une majorité de 0 seront transformées en variables binaires, pour ne pas fausser les prédictions.

```python
quantiles = [i/10 for i in range(1, 10)]

for col in list(df_num.columns.values):
    if(np.quantile(df_num[col], 0.25) == np.quantile(df_num[col], 0.75)):
        print(df_num[col].describe(percentiles=quantiles))
```

```python
col_bool = ["BsmtFinSF2", "LowQualFinSF", "BsmtHalfBath", "EnclosedPorch", "3SsnPorch",
           "ScreenPorch", "PoolArea", "MiscVal"]
```

```python
for col in col_bool:
    data[col] = (data[col] != 0)
```

```python
# DEDAWP
df_num = data.select_dtypes(include = ['float64', 'int64'])
for i in range(0, len(df_num.columns), 5):
    sns.pairplot(data=df_num,
                x_vars=df_num.columns[i:i+5],
                y_vars=['SalePrice'])
```

### On peut voir quelles variables semblent corrélées ; nous pouvons espérer les retrouver dans notre analyse finale.

### Observons les relations entre SalePrice et les variables qualitatives

```python
df_qual = data.select_dtypes(exclude=['float64', 'int64'])

for col in list(df_qual.columns.values):
    data.boxplot(column="SalePrice", by=col)
```

### Certaines variables qualitatives ont trop de classes. Voyons lesquelles

```python
for col in list(df_qual.columns.values):
    if (len(df_qual[col].value_counts())>5):
        print(df_qual[col].value_counts())
```

```python
# Commençons par regrouper les valeurs rares :

for col in list(df_qual.columns.values):
    for val in data[col].value_counts().index[data[col].value_counts()<30]:
        data[col][data[col]==val] = 'other'
```

```python
df_qual = data.select_dtypes(exclude=['float64', 'int64'])
trop_longues = []

for col in list(df_qual.columns.values):
    if (len(df_qual[col].value_counts())>5):
        print(df_qual[col].value_counts())
        trop_longues.append(col)
```

```python
# Séparons les données d'entraînement et de validation
train, test = train_test_split(data, test_size=0.3, random_state=58)
```

```python
cat = df_qual.columns.values
target_encoder = TargetEncoder()
train[cat] = target_encoder.fit_transform(train[cat], train.SalePrice)
test[cat] = target_encoder.transform(test[cat])
```

```python
# DELETE
# Nous transformerons les variables trop longues restantes en variables numériques
target_encoder = TargetEncoder()
train[trop_longues] = target_encoder.fit_transform(train[trop_longues], train.SalePrice)
test[trop_longues] = target_encoder.transform(test[trop_longues])
```

```python
train.to_csv("train.csv")
test.to_csv("test.csv")
```
